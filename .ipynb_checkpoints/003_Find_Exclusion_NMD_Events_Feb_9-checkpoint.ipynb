```python
database='/Users/rhythmicstar/projects/exon_evolution//gencode.v19.annotation.outrigger.gtf.db'
nmd_file='/Users/rhythmicstar/ipython_notebook/nmd/example_splicing.bed'
event_ids_df = nmd.possible_nmd(nmd_file)
DBHuman = gffutils.FeatureDB(database)
```

```python
def trans_exon_set(exon):
    transcripts_with_exon = set()
    for exon_transc in DBHuman.parents(exon, featuretype="transcript"):
        transcripts_with_exon.add(exon_transc['transcript_id'][0])
    return transcripts_with_exon #parent_transcripts_of_exon

def all_trans_set(exon):
    all_transcripts = set()
    for transc in DBHuman.children(exon['gene_id'][0], featuretype="transcript"):
        all_transcripts.add(transc['transcript_id'][0])
    return all_transcripts #call it transcripts_from_gene_containing_exon

def create_dict(all_transcripts, strand):
    transcript_exons = dict()
    for transcript_id in all_transcripts:
        this_transcript_exons = tuple(DBHuman.children(transcript_id, featuretype = "exon", order_by="start"))
        #set reverse (look up) = true
        internal_exons = this_transcript_exons[1:len(this_transcript_exons)-1]
        transcript_cds = tuple(DBHuman.children(transcript_id, featuretype='CDS', order_by="start"))
        for startcodon in DBHuman.children(transcript_id, featuretype='start_codon'):
            if (strand=='+'):
                firstcds = transcript_cds[0:1]
                real_transcript_cds = ((startcodon,) + (firstcds) + internal_exons)
            else:
                firstcds = transcript_cds[-1]
                real_transcript_cds = (internal_exons + (firstcds,) + (startcodon,)) #can use reverse order of internal exons
            #print("real_transcript_cds", real_transcript_cds)
            transcript_exons[transcript_id] = real_transcript_cds
    return transcript_exons

    
def inclusion_possible_nmd(trans_without_exon, trans_with_exon, strand):
    trans_with_exon_set = set(trans_with_exon)
    trans_without_exon_set = set(trans_without_exon)
    as_exon = tuple(trans_with_exon_set-trans_without_exon_set)[0]
    #print(as_exon)
    index = trans_with_exon.index(as_exon)
    if (strand=='+'):
        for iterator in range(index, len(trans_with_exon)):
            for stop in DBHuman.children(trans_with_exon[iterator]['gene_id'][0], featuretype='stop_codon'):
                #print("stop codon: ", stop)
                if stop_codon_nmd(trans_with_exon[index], stop):
                    return True
    if (strand=='-'):
        for iterator in range(0, index+1):
            for stop in DBHuman.children(trans_with_exon[iterator]['gene_id'][0], featuretype='stop_codon'):
                #print("stop codon: ", stop)
                if stop_codon_nmd(trans_with_exon[index], stop):
                    return True
    return False


def exclusion_possible_nmd(trans_without_exon, trans_with_exon, strand):
    with open("without_exon.gtf", "w") as f:
        for line in trans_without_exon:
            f.write(str(line))
    with open("with_exon.gtf", "w") as f:
        for line in trans_with_exon:
            f.write(str(line))
    print("trans_with_exon: ", trans_with_exon)
    print("trans_without_exon: ", trans_without_exon)
    trans_with_exon_set = set(trans_with_exon)
    trans_without_exon_set = set(trans_without_exon)
    as_exon = tuple(trans_with_exon_set-trans_without_exon_set)[0]
    print("as_exon: ", as_exon)
    #print(as_exon)
    index = trans_with_exon.index(as_exon)
    if (strand=='+'):
#       print("trans_without_exon: ", trans_without_exon)
        for iterator in range(index, len(trans_without_exon)+1):
            for stop in DBHuman.children(trans_without_exon[iterator]['gene_id'][0], featuretype='stop_codon'):
#               print("stop codon: ", stop)
                if stop_codon_nmd(trans_without_exon[index], stop):
                    return True
    else:
        for iterator in range(0, index):
            for stop in DBHuman.children(trans_with_exon[iterator]['gene_id'][0], featuretype='stop_codon'):
                #print("stop codon: ", stop)
                if stop_codon_nmd(trans_with_exon[index], stop):
                    return True
    return False
    
def stop_codon_nmd(human_exon, stop_codon):
    if stop_codon.strand == '+':
        if stop_codon.start >= human_exon.start and stop_codon.end <= \
                human_exon.end and stop_codon.end + 50 < human_exon.end:
            return True
    else:
        if stop_codon.start >= human_exon.start and stop_codon.end <= \
                human_exon.end and human_exon.start + 50 < stop_codon.start:
            return True
    return False
```

```python
for i, row in event_ids_df.iterrows():
    print('-----------------')
    print(row['exon'])
    inclusion_exons = set()
    exclusion_exons = set()
    location = 'exon:' + row['exon'] + ":" + row['strand']
    print(location)
    try:
        exon = DBHuman[location]
        transcripts_with_exon = trans_exon_set(exon)
        all_transcripts = all_trans_set(exon)
        transcripts_without_exon = all_transcripts-transcripts_with_exon
        transcript_exons = create_dict(all_transcripts, row['strand'])

        for t1 in transcripts_without_exon:
            try:
                for t2 in transcripts_with_exon:
                    t1_ex = transcript_exons[t1]
                    t2_ex = transcript_exons[t2]
#                     print("t1", t1_ex)
                    if ((row['strand']=='+' and t1_ex[1] == t2_ex[1]) or (row['strand']=='-' and t1_ex[-1] == t2_ex[-1])):
                        if len(set(t2_ex).symmetric_difference(set(t1_ex)))==1:
                            if (DBHuman[t1]['transcript_type'][0]!='nonsense_mediated_decay') & (DBHuman[t2]['transcript_type'][0]=='nonsense_mediated_decay'):
                                if inclusion_possible_nmd(t1_ex, t2_ex, row['strand']):
                                    inclusion_exons.add(t2)
                            if (DBHuman[t1]['transcript_type'][0]=='nonsense_mediated_decay') & (DBHuman[t2]['transcript_type'][0]!='nonsense_mediated_decay'):
                                if exclusion_possible_nmd(t1_ex, t2_ex, row['strand']):
                                    print("True")
                                    exclusion_exons.add(t1)
                            print("inc", inclusion_exons)
                            print("exc", exclusion_exons)

            except KeyError:
                t1_ex=None
    except gffutils.FeatureNotFoundError:
        exon = None
```
