```python
#for pandas data frame
import pandas as pd
```

```python
# splicing_data = pd.read_csv('/Users/rhythmicstar/ipython_notebook/nmd/correlated_actual_filtered.csv')
# splicing_data = pd.read_csv('/Users/rhythmicstar/ipython_notebook/nmd/events_beyond_cutoff.csv')
splicing_data = pd.read_csv('/Users/rhythmicstar/ipython_notebook/nmd/events_beyond_cutoff_ipsc_npc_mn.csv')

index = pd.Index(splicing_data['event_id'])
event_ids = pd.Series(index, name='event_id')
event_ids_df = event_ids.to_frame()
event_ids_df['exon'] = splicing_data['event_id'].map(lambda x: x.split('@')[1])
event_ids_df.drop(['event_id'],inplace=True,axis=1)
event_ids_df.drop_duplicates(inplace=True)
event_ids_df = event_ids_df[event_ids_df.exon.str.contains("novel") == False]
# event_ids_df['exon'].to_csv('correlated_actual_filtered_targets.csv', index=False)
# event_ids_df['exon'].to_csv('events_beyond_cutoff_targets.csv', index=False)
event_ids_df['exon'].to_csv('events_beyond_cutoff_ipsc_npc_mn_targets.csv', index=False)
```
